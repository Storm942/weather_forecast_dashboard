```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Load the dataset (simulate real-time data source here)
data = pd.read_csv('weather_data.csv')  # Replace with real-time data if available

# Convert 'Date_Time' to datetime format and set as index
data['Date_Time'] = pd.to_datetime(data['Date_Time'])
data.set_index('Date_Time', inplace=True)

# Check for missing values
print("Missing Values:\n", data.isnull().sum())

# Fill missing values with forward fill method
#data.fillna(method='ffill', inplace=True)
data.ffill(inplace=True)  # Forward fill

# Use forward fill or backward fill as needed
#data.bfill(inplace=True)  # Backward fill


# Select features (you can add more relevant features as necessary)
features = ['Temperature_C', 'Humidity_pct', 'Precipitation_mm', 'Wind_Speed_kmh']
data = data[features]

# Scale the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)


# Set path chunksize to handle more complex plots
plt.rcParams['agg.path.chunksize'] = 10000  # Adjust this as needed

# Optionally, adjust the path simplification threshold to reduce complexity
plt.rcParams['path.simplify_threshold'] = 0.1  # Lower this to simplify further


# Plot initial data for trends
plt.figure(figsize=(10, 6))
plt.plot(data['Temperature_C'], label='Temperature (C)', color='blue')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Function to create sequences for time series data
def create_sequences(data, look_back, forecast_horizon=1):
    X, y = [], []
    for i in range(len(data) - look_back - forecast_horizon + 1):
        X.append(data[i:i + look_back, :])
        y.append(data[i + look_back:i + look_back + forecast_horizon, 0])  # Forecast the temperature
    return np.array(X), np.array(y)

# Define look-back (historical days) and forecast horizon (days to predict)
look_back = 7  # Example: using 7 days of data for prediction
forecast_horizon = 5  # Predict the next 5 days

# Create sequences
X, y = create_sequences(scaled_data, look_back, forecast_horizon)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print("Shape of Training Data:", X_train.shape)
print("Shape of Testing Data:", X_test.shape)
```

```python
# Import necessary libraries
import keras_tuner as kt

from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam


# Define a hypermodel for Keras Tuner
def build_model(hp):
    model = Sequential()
    
    # Explicit Input layer
    model.add(layers.Input(shape=(look_back, X.shape[2])))
    
    # First LSTM layer with hyperparameter tuning for the number of units
    model.add(layers.LSTM(units=hp.Int('units', min_value=32, max_value=128, step=32),
                          activation='relu', return_sequences=True))
    model.add(layers.Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))

    # Additional LSTM layer
    model.add(layers.LSTM(units=hp.Int('units_2', min_value=32, max_value=128, step=32),
                          activation='relu', return_sequences=False))
    model.add(layers.Dropout(hp.Float('dropout_2', 0.1, 0.5, step=0.1)))

    # Output layer
    model.add(layers.Dense(forecast_horizon))  # Forecast multiple days

    # Compile model
    model.compile(
        optimizer=Adam(hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='LOG')),
        loss='mse',
        metrics=['mae']
    )
    return model

# Initialize Keras Tuner
tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,
    executions_per_trial=1,
    directory='tuner1_dir',
    project_name='weather_forecast_lstm'
)

# Run the hyperparameter search
tuner.search(X_train, y_train, epochs=50, validation_data=(X_test, y_test), batch_size=16)
```

```python
# Import necessary libraries
import keras_tuner as kt
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

# Reinitialize the tuner to resume from saved progress
tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,
    executions_per_trial=1,
    directory='tuner1_dir',
    project_name='weather_forecast_lstm'
)

# Resume the search
tuner.search(X_train, y_train, epochs=50, validation_data=(X_test, y_test), batch_size=16)

# Print results summary
tuner.results_summary()
```

```python
# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]

# Evaluate the best model on test set
test_loss, test_mae = best_model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}, Test MAE: {test_mae}")

# Make predictions
y_pred = best_model.predict(X_test)

# Rescale predictions
y_pred_rescaled = scaler.inverse_transform(
    np.c_[y_pred[:, 0], np.zeros((len(y_pred), len(features)-1))]
)[:, 0]

y_test_rescaled = scaler.inverse_transform(
    np.c_[y_test[:, 0], np.zeros((len(y_test), len(features)-1))]
)[:, 0]

# Plot Actual vs Predicted Temperature
plt.figure(figsize=(10, 6))
plt.plot(y_test_rescaled, label='Actual Temperatures')
plt.plot(y_pred_rescaled, label='Predicted Temperatures', linestyle='dashed')
plt.xlabel('Time Steps')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Function to forecast future temperatures
def forecast_future(model, data, scaler, look_back, forecast_horizon):
    predictions = []
    input_seq = data[-look_back:, :]
    
    for _ in range(forecast_horizon):
        input_seq_reshaped = input_seq.reshape(1, look_back, len(features))
        pred = model.predict(input_seq_reshaped)
        predictions.append(pred[0, 0])
        
        # Update input sequence
        next_input = np.append(pred[0, 0], input_seq[-1, 1:]).reshape(1, -1)
        input_seq = np.vstack([input_seq[1:], next_input])
    
    # Rescale predictions
    predictions_rescaled = scaler.inverse_transform(
        np.c_[predictions, np.zeros((len(predictions), len(features)-1))]
    )[:, 0]
    return predictions_rescaled

# Predict for the next 7 days
future_days = 7
future_predictions = forecast_future(best_model, scaled_data, scaler, look_back, future_days)

print(f"Predicted Temperatures for the Next {future_days} Days: {future_predictions}")

# Plot future predictions
plt.figure(figsize=(8, 4))
plt.plot(range(1, future_days + 1), future_predictions, marker='o', color='red', label='Future Predictions')
plt.title(f'Temperature Forecast for the Next {future_days} Days')
plt.xlabel('Days Ahead')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```
