```python
# Import necessary libraries
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler

```

```python
# Load the dataset
data = pd.read_csv('weather_data.csv')  # Replace with your dataset path
data['Date_Time'] = pd.to_datetime(data['Date_Time'])
data.set_index('Date_Time', inplace=True)

# Select features for model prediction
features = ['Temperature_C', 'Humidity_pct', 'Precipitation_mm', 'Wind_Speed_kmh']
data = data[features]

```

```python
# Load the previously saved scaler
scaler = joblib.load('scaler.pkl')  # Load the scaler object used during training

# Scale the new data (use transform instead of fit_transform)
scaled_data = scaler.transform(data)  # Scale the data with the pre-trained scaler

# Load the saved model
model = load_model('weather_forecast_model_new.keras')  # Load the best trained model

print("Model loaded successfully!")

```

```python
# Define the function to forecast with uncertainty
def forecast_with_uncertainty(model, data, scaler, look_back, forecast_horizon, n_simulations=100):
    predictions = []
    input_seq = data[-look_back:, :]
    
    for _ in range(forecast_horizon):
        input_seq_reshaped = input_seq.reshape(1, look_back, len(features))
        simulated_preds = []
        
        # Run multiple simulations for uncertainty
        for _ in range(n_simulations):
            pred = model.predict(input_seq_reshaped, verbose=0)
            simulated_preds.append(pred[0, 0])
        
        # Calculate mean and standard deviation
        mean_pred = np.mean(simulated_preds)
        std_pred = np.std(simulated_preds)
        
        predictions.append((mean_pred, std_pred))
        
        # Update the input sequence
        next_input = np.append(mean_pred, input_seq[-1, 1:]).reshape(1, -1)
        input_seq = np.vstack([input_seq[1:], next_input])
    
    # Rescale predictions
    predictions_rescaled = scaler.inverse_transform(
        np.c_[[p[0] for p in predictions], np.zeros((len(predictions), len(features)-1))]
    )[:, 0]
    
    intervals = [1.96 * p[1] for p in predictions]  # 95% Confidence Interval
    return predictions_rescaled, intervals



```

```python
# Predict and plot for the next 7 days
future_days = 7
future_predictions, confidence_intervals = forecast_with_uncertainty(
    model, scaled_data, scaler, look_back=7, forecast_horizon=future_days
)

# Plot predictions with confidence intervals
plt.figure(figsize=(8, 4))
plt.plot(range(1, future_days + 1), future_predictions, marker='o', color='red', label='Future Predictions')
plt.fill_between(range(1, future_days + 1),
                 future_predictions - confidence_intervals,
                 future_predictions + confidence_intervals,
                 color='gray', alpha=0.3, label='Confidence Interval')
plt.title(f'Temperature Forecast for the Next {future_days} Days')
plt.xlabel('Days Ahead')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Load and preprocess the dataset again if necessary
data = pd.read_csv('weather_data.csv')  # Load data again if necessary
data['Date_Time'] = pd.to_datetime(data['Date_Time'])
data.set_index('Date_Time', inplace=True)

# Select features and scale the data
features = ['Temperature_C', 'Humidity_pct', 'Precipitation_mm', 'Wind_Speed_kmh']
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data[features])

# Function to create sequences
def create_sequences(data, look_back, forecast_horizon=1):
    X, y = [], []
    for i in range(len(data) - look_back - forecast_horizon + 1):
        X.append(data[i:i + look_back, :])
        y.append(data[i + look_back:i + look_back + forecast_horizon, 0])  # Forecast the temperature
    return np.array(X), np.array(y)

# Set look_back and forecast_horizon
look_back = 7  # Using 7 days of data for prediction
forecast_horizon = 5  # Predict the next 5 days

# Create sequences
X, y = create_sequences(scaled_data, look_back, forecast_horizon)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Check the shape of X_test to ensure it's correctly defined
print("Shape of X_test:", X_test.shape)

```

```python
# Load the previously saved model
best_model = load_model('weather_forecast_model_new.keras')

# Assuming the scaler was saved earlier and is available
import joblib
scaler = joblib.load('scaler.pkl')

# Make predictions
y_pred = best_model.predict(X_test)

# Rescale predictions and actual values to the original scale
y_pred_rescaled = scaler.inverse_transform(
    np.c_[y_pred[:, 0], np.zeros((len(y_pred), len(features)-1))]  # Assuming the target is temperature
)[:, 0]

y_test_rescaled = scaler.inverse_transform(
    np.c_[y_test[:, 0], np.zeros((len(y_test), len(features)-1))]
)[:, 0]

# Calculate RMSE and R²
from sklearn.metrics import mean_squared_error, r2_score

rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))
r2 = r2_score(y_test_rescaled, y_pred_rescaled)

print(f"RMSE: {rmse}")
print(f"R²: {r2}")

```

```python
# Adjusted function to forecast future temperatures
def forecast_future(model, data, scaler, look_back, forecast_horizon):
    predictions = []
    input_seq = data[-look_back:, :]
    
    for _ in range(forecast_horizon):
        input_seq_reshaped = input_seq.reshape(1, look_back, len(features))
        pred = model.predict(input_seq_reshaped, verbose=0)
        predictions.append(pred[0, 0])  # Store temperature prediction
        
        # Update the input sequence
        next_input = np.append(pred[0, 0], input_seq[-1, 1:]).reshape(1, -1)
        input_seq = np.vstack([input_seq[1:], next_input])
    
    # Rescale predictions to original scale
    predictions_rescaled = scaler.inverse_transform(
        np.c_[predictions, np.zeros((len(predictions), len(features) - 1))]
    )[:, 0]
    return predictions_rescaled

# Forecast for the next 7 days
future_days = 7
look_back = 7  # Using the same look-back window as in training
future_predictions = forecast_future(best_model, scaled_data, scaler, look_back, future_days)

print(f"Predicted Temperatures for the Next {future_days} Days: {future_predictions}")

# Plot future predictions
plt.figure(figsize=(8, 4))
plt.plot(range(1, future_days + 1), future_predictions, marker='o', color='red', label='Future Predictions')
plt.title(f'Temperature Forecast for the Next {future_days} Days')
plt.xlabel('Days Ahead')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

def plot_forecast(future_predictions, confidence_intervals, future_days):
    """
    Plots the temperature forecast for the next 'future_days' days along with the confidence interval.
    
    Parameters:
        future_predictions (list or np.array): Forecasted temperature values for the next 'future_days' days.
        confidence_intervals (list or np.array): The corresponding confidence intervals for the forecast.
        future_days (int): The number of future days for which the forecast is made.
    """
    plt.figure(figsize=(8, 4))
    
    # Plot the future predictions
    plt.plot(range(1, future_days + 1), future_predictions, marker='o', color='red', label='Future Predictions')
    
    # Plot the confidence intervals
    plt.fill_between(range(1, future_days + 1),
                     future_predictions - confidence_intervals,
                     future_predictions + confidence_intervals,
                     color='gray', alpha=0.3, label='Confidence Interval')
    
    # Add title, labels, and legend
    plt.title(f'Temperature Forecast for the Next {future_days} Days')
    plt.xlabel('Days Ahead')
    plt.ylabel('Temperature (°C)')
    plt.legend()
    
    # Show the plot
    plt.show()

```

```python
future_days = 7
future_predictions = forecast_future(model, scaled_data, scaler, look_back=7, forecast_horizon=future_days)

# For simplicity, let's assume a constant confidence interval (you can modify as needed)
confidence_intervals = [1.96 * 0.5] * len(future_predictions)  # 95% confidence interval, assume std = 0.5

# Call the plot function to visualize the forecast
plot_forecast(future_predictions, confidence_intervals, future_days)

```

```python
# Install dash if not already installed
# !pip install dash

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the dashboard
app.layout = html.Div([
    html.H1("Weather Forecast Dashboard", style={'text-align': 'center'}),
    
    dcc.Graph(id='historical-trend'),
    
    html.Div([
        html.Label("Days to Forecast:"),
        dcc.Slider(
            id='forecast-slider',
            min=1,
            max=14,
            step=1,
            marks={i: str(i) for i in range(1, 15)},
            value=7,
        )
    ], style={'width': '50%', 'margin': 'auto'}),
    
    dcc.Graph(id='forecast-trend')
])

# Callback to update graphs based on user input
@app.callback(
    [Output('historical-trend', 'figure'),
     Output('forecast-trend', 'figure')],
    [Input('forecast-slider', 'value')]
)
def update_graphs(days_to_forecast):
    # Historical data
    historical_fig = go.Figure()
    historical_fig.add_trace(go.Scatter(
        x=data.index, y=data['Temperature_C'],
        mode='lines', name='Temperature'
    ))
    historical_fig.update_layout(title="Historical Temperature Trend", xaxis_title="Date", yaxis_title="Temperature (C)")
    
    # Forecast future predictions
    future_preds = forecast_future(model, scaled_data, scaler, look_back, days_to_forecast)
    forecast_fig = go.Figure()
    forecast_fig.add_trace(go.Scatter(
        x=list(range(1, days_to_forecast + 1)),
        y=future_preds,
        mode='lines+markers', name='Predicted Temperature'
    ))
    forecast_fig.update_layout(title=f"Forecast for Next {days_to_forecast} Days", xaxis_title="Days Ahead", yaxis_title="Temperature (C)")
    
    return historical_fig, forecast_fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)

streamlit run "c:\jupyter notebook files\weather_forecast_dashboard.py"
```
